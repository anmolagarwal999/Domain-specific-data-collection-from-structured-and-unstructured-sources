```python
import os
import json
```

```python
FILE_PATH="./CNT_gs_viaf_dblp_wc_mgp_wikiquote_influ_semantic_research.json"
with open(FILE_PATH, 'r') as fd:
    df=json.load(fd)
```

```python
len(df)
```

```python
num_cols=[]
```

```python
for curr_elem in df:
    num=len(curr_elem.keys())
    num_cols.append(num)
```

```python
len(num_cols)
```

```python
# import libraries
from matplotlib import pyplot as plt 
import numpy as np  
```

# Plot 1

```python
fig = plt.figure(figsize =(10, 7))
  
plt.hist(num_cols) 
plt.title("Number of attributes per row",fontsize=22)
plt.xlabel("Number of attributes",fontsize=16)
plt.ylabel("Number of rows having 'x' attributes",fontsize=16)
```
