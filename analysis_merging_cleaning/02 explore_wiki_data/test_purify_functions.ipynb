```python
def find_relevant_func(key_name):
    for idx, elem in enumerate(master_mapping):
        if elem['attr_name']==key_name:
            return elem['func']
    assert(False)
    return None
```

```python
def find_relevant_key(key_name):
    for idx, elem in enumerate(master_mapping):
        if elem['attr_name']==key_name:
            if elem['parent_attr']=="":
                return key_name
            else:
                return elem['parent_attr']
    assert(False)
    return None
```

```python
from copy import deepcopy
```

```python
new_df=[]
```

```python
for curr_elem in df:
    new_elem=dict()
    for curr_key in curr_elem:
        f_to_use=find_relevant_func(curr_key)
        if f_to_use!=None:
            #print(curr_key)
            #print(curr_elem[curr_key])
            curr_elem[curr_key]=f_to_use(curr_elem[curr_key])
        main_key=find_relevant_key(curr_key)
        if main_key in new_elem:
            if type(curr_elem[curr_key])!=list:
                    curr_elem[curr_key]=[curr_elem[curr_key]]
            if type(new_elem[main_key])==list:
                new_elem[main_key].extend(deepcopy(curr_elem[curr_key]))
            else:
                new_elem[main_key]=[new_elem[main_key]]
                new_elem[main_key].extend(deepcopy(curr_elem[curr_key]))                
        else:
            new_elem[main_key]=deepcopy(curr_elem[curr_key])
        if type(new_elem[main_key])==list:
            try:
                new_elem[main_key]=list(set(new_elem[main_key]))
            except:
                pass
    new_df.append(deepcopy(new_elem))
```
