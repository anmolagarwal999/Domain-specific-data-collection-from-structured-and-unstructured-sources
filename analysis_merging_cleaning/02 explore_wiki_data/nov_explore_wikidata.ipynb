```python
#### FInd frequency of top level attributes
```

```python
sample_top={}
```

```python
top_freq={}
for curr_elem in df:
    for top_key in curr_elem.keys():
        try:
            top_freq[top_key]+=1
        except:
            sample_top[top_key]=curr_elem[top_key]
            top_freq[top_key]=1
print("Top freq is ")
print(*(top_freq.items()), sep='\n')
            
```

```python
def purify_backlinks(arr):
    return [x['title'] for x in arr]
```

```python
def purify_image(arr):
    if type(arr)!=list:
        #print(arr)
        return None
    return list(set([x['url'] for x in arr]))
```

```python
this_set=set()
```

```python
for curr_elem in df:
    keys_list=list(curr_elem.keys())
    for curr_key in keys_list:
        if curr_key=="image":
            use_val=purify_image(curr_elem[curr_key])
            assert(use_val!=0)
            if use_val==None:
                del curr_elem[curr_key]
            else:
                curr_elem[curr_key]=use_val
                
        elif  curr_key=="backlinks":
            curr_elem[curr_key]=purify_backlinks(curr_elem[curr_key])
        else:
            if type(curr_elem[curr_key])==dict:
                for this_key, this_val in curr_elem[curr_key].items():
                    curr_elem[this_key]=this_val
                del curr_elem[curr_key]

```

```python
for curr_elem in df:
    keys_list=list(curr_elem.keys())
    for curr_key in keys_list:
        if curr_elem[curr_key]==None:
            del curr_elem[curr_key]
```

```python
bottom_freq={}
for curr_elem in df:
    for bottom_key in curr_elem.keys():
        try:
            bottom_freq[bottom_key]+=1
        except:
            sample_top[bottom_key]=curr_elem[bottom_key]
            bottom_freq[bottom_key]=1
print("Bottom freq is ")
print(*(bottom_freq.items()), sep='\n')
            
```

```python
sorted_freq_list=sorted(list(bottom_freq.items()),key=lambda x:-x[1])
```
